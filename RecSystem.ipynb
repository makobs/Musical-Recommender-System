```python
import pandas as pd
import numpy as np
```

```python
#all the data
df = pd.read_csv("m4allSmall/listening_history.csv", sep = "\t") #listening_history
genres = pd.read_csv("m4allSmall/id_genres.csv", sep = "\t") #id_genres.csv
information = pd.read_csv("m4allSmall/id_information.csv", sep = "\t") #id_information.csv
lang = pd.read_csv("m4allSmall/id_lang.csv", sep = "\t") #id_lang.csv
metadata = pd.read_csv("m4allSmall/id_metadata.csv", sep = "\t") #id_metadata.csv
tags = pd.read_csv("m4allSmall/id_tags.csv", sep = "\t") #id_tags.csv
```

```python
df = df.rename(columns = {"song": "id"})
df.head() #5109592 rows × 3 columns
```

```python
full = df.merge(information, how = 'outer')
full = full.merge(tags, how = 'outer')
full = full.merge(genres, how = 'outer')
full = full.merge(lang, how = 'outer')
full = full.merge(metadata, how = 'outer')

full = full.rename(columns = {"id": "song_id"})
full.head() #5119265 rows × 19 columns
```

```python
full.describe()
```

```python
full.info()
```

```python
full.dropna(inplace = True) #get rid of nas
# full.isna().sum()
full.shape #(5109592, 19)
```

```python
#where user & song_id == same, merge, make song count
songCount = full.groupby(['user', 'song_id']).size().reset_index(name = 'song_count')
full = full.merge(songCount, how = 'outer')
full = full.drop_duplicates(subset = ['user', 'song_id'], keep='last').reset_index(drop = True)

full.head() #2597382 rows × 20 columns
```

```python
#the dataset to train the recommender system on
simple = pd.concat([full['user'], full['song_id'], full['song_count']], axis = 1)
simple.describe()

simple
```

---

```python
#dictionary of song_id to song & artist
songID = information.loc[:, 'id']
songName = information.loc[:, 'song']
songArtist = information.loc[:, 'artist']

songDict = {z[0]: list(z[1:]) for z in zip(songID, songName, songArtist)}
songDict
```

```python
#creating count matrix from song_counts
countMatrix = simple.pivot(index = 'user', columns = 'song_id', values = 'song_count')
userID = countMatrix.index
songID = countMatrix.columns


#can't use nan to train model replace with 0
#creation of the sparse matrix
countMatrix.fillna(0, inplace = True)
countMatrix = countMatrix.astype(np.int32)

#transpose so it is songs x users
countMatrixPivot = countMatrix.T
countMatrixPivot.head()
```

```python
#function to get songID from songName & songArtist
def correspondingKey(sName, sArtist):
    for key, value in songDict.items():
        if (sName and sArtist) in value:
            return key
    return None


#function to return recommended songs as the name and artist, not id
def idToName(idMatrix):
    return [songDict.get(key) for key in idMatrix]
```

## KNN Recommender Approach

```python
from sklearn.neighbors import NearestNeighbors
```

```python
metric = 'cosine'
algorithm = 'brute'
k = 5 #number of recommendations

modelKnn = NearestNeighbors(metric = metric, algorithm = algorithm)
modelKnn.fit(countMatrixPivot.values)
```

```python
#function to return recommended songs 
#re: Jenny Tang
def makeRecKNN(k, sName, sArtist):
    try: 
        songKey = correspondingKey(sName, sArtist)
        song = countMatrixPivot.loc[songKey]
    except KeyError as e:
        print('Cannot find the song', e, 'in database!')
        return
    
    similarities = []
    indices = []
    
    distance, indice = modelKnn.kneighbors([song.values], n_neighbors = k + 1)
    
    songNamesArtist = np.array(idToName(countMatrixPivot.iloc[indice[0]].index.values))
    songNames = songNamesArtist[:, 0]
    songArtists = songNamesArtist[:, 1]
    
    similarities = 1 - distance.flatten()
    
    recommended_books = pd.DataFrame({
        'Song Name'   : songNames,
        'Song Artist' : songArtists,
        'similarities': similarities}).sort_values(by = 'similarities', ascending = False)

    return recommended_books
```

```python
makeRecKNN(5, 'Burn It To The Ground', 'Nickelback')
```
